```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import *
from pymatgen.analysis.diffraction.xrd import *
from matplotlib import pyplot
```

```python
# Obtenir l'objet structure du CdCl2
with MPRester("sdp42oAyhJ8t1iJnx") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-22881")
    w = CifWriter(structure, 0.1)
    w.write_file('mp-22881.cif')
```

```python
# détermine les pics pour un rayonnement X incident Cu Ka
xrd = XRDCalculator(wavelength='CuKa', symprec=0, debye_waller_factors=None)
pattern = xrd.get_pattern(structure, scaled=True, two_theta_range=(0, 90))
```

```python
# détermine les indices hkl associés
hkl = pattern.hkls
```

```python
# affiche le diffractogramme pour un rayonnement X incident Cu Kα
xrd.show_plot(structure, annotate_peaks=False)
```

```python
# affiche les données des pics
print('Données du premier pics :')
print('     indice hkl :', hkl[0][0]["hkl"], 'de multiplicité', hkl[0][0]["multiplicity"])
print('     coordonnées (x,y) = (', pattern.x[0], ',', pattern.y[0], ')')
print('Données du second pics :')
print('     indice hkl :', hkl[1][0]["hkl"], 'de multiplicité', hkl[1][0]["multiplicity"])
print('     coordonnées (x,y) = (', pattern.x[1], ',', pattern.y[1], ')')
print('Données du troisième pics :')
print('     indice hkl :', hkl[2][0]["hkl"], 'de multiplicité', hkl[2][0]["multiplicity"])
print('     coordonnées (x,y) = (', pattern.x[2], ',', pattern.y[2], ')')
```
